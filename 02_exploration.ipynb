## Visualisation du dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cr√©ation du dataframe
df = pd.read_csv("assurance_dataset_clean.csv")
df.head(5)
```

```python
df.info()
```

```python
# Visualisation globale des donn√©es
sns.pairplot(data=df)
plt.show()
```

```python
# Visualisation des donn√©es en fonction du sexe
sns.pairplot(data=df, hue='sex')
plt.show()
```

```python
# Visualisation des donn√©es en fonction da qualit√© de fumeur
sns.pairplot(data=df, hue='smoker')
plt.show()
```

```python
# Visualisation des donn√©es en fonction des r√©gions
sns.pairplot(data=df, hue='region')
plt.show()
```
