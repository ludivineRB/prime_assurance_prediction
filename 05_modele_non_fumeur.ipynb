## Prédictions de la prime d'assurance pour les non-fumeurs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
%matplotlib inline
```

```python
# Création du dataset "non-fumeur"
df = pd.read_csv('assurance_dataset_clean.csv')
df_no_fumeur = df[(df['smoker'] == False)]
df_no_fumeur.head(5)
```

```python
# préparation des données d'entrainement
X = df_no_fumeur[['age','bmi']].values 
Y = df_no_fumeur['charges'].values

print(X[:5])
print(Y[:5])
```

```python
modele = linear_model.LinearRegression()
modele.fit(X,Y)
```

```python
print("coef =", modele.coef_)
print("b =", modele.intercept_)
```

```python
# Calcul des prédictions

Y_pred = modele.predict(X)
print("Y ", Y[:10])
print("Y_pred", Y_pred[:10])
```

```python
# Visualisation
plt.plot(Y[:100], label='charges')
plt.plot(Y_pred[:100], label = 'prédictions')
plt.legend()
plt.show()
```
