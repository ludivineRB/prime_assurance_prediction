```python
#Import

import pandas as pd
import missingno as msno
```

## Importation of the datas and first visualization

```python
df = pd.read_csv('assurance_data.csv', sep=',')
df.head()
```

```python
df.info()
```

```python
df.isnull().head()
```

```python
#Count of null data
df.isnull().sum()
```

## Test missingno

```python
msno.matrix(df)
```

```python
msno.bar(df)
```

```python
msno.heatmap(df)
```

## Checking for duplicates

```python
df.duplicated().count()
df.duplicated().sum()
df.duplicated().value_counts()
```

```python
#Deletion of duplicates
df.drop_duplicates(inplace=True)
```
